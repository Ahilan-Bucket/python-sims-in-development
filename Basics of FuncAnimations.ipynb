Basics of Plot Animation

Here Build a Sine wave Animation

```python
# Basic Libraries

import numpy as np
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
```

Let's Build the Sin Wave Data for this Graph which is easy

```python
# we want data that goes from x [0, 10pi]

x = np.arange(0,10*np.pi,0.01) # 0.01 is the step space between each interval
y = np.sin(x) # There, our sin wave
```

Let's Build our Basic Canvas

```python
fig = plt.figure()
ax = plt.subplot(1,1,1)
```

These have way too Many Data Points so lets skip some data

```python
n_skip = 50
```

Let's Build an Inital State Function that tells the pen where to start.
This is called when ever the animation starts and when ever it repeats

```python
def init():
    ax.clear()
    plt.xlabel(r"$\pi$")
    plt.ylabel(r"$Sin(\pi)$")
    plt.xlim(x[0],max(x))
    plt.ylim(min(y),max(y))
```

Lets make an Update function that first takes the Inital Pen parameters and changes it.

This is called repeatedly by the FuncAnimation object


```python
def update(i):
    ax.plot(x[i:i+n_skip],y[i:i+n_skip],color='k')
    ax.scatter(x[i],y[i],marker='o',color='b')
```

Now Finally the code that uses and calls these two Functions, the FuncAnimation

```python
animate = FuncAnimation(
    fig,# Figure Obejct Needed to do process
    func = update,# Function to call at each frame
    frames = np.arange(0,len(x),n_skip), # Data to pass into func at every iteration or frame
    # If int, Matplotlib itself runs a loop until int-1.
    init_func=init, # Function to call at first and every repeated animation. TO clear animation
    interval=20, # Delay in ms between the speed of each rendertion(Does not change Physics), just for eyes to catchup
)
```

Finally Show The Animation

```python
from IPython.display import HTML
plt.rcParams["animation.html"] = "jshtml"
HTML(animate.to_jshtml())
```
